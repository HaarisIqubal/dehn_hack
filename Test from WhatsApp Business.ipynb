```python
from sentence_transformers import SentenceTransformer
import numpy as np
import torch

# Pick a model
model_name = "sentence-transformers/all-MiniLM-L6-v2"  # fast, 384-dim

# Let torch detect GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the model
model = SentenceTransformer(model_name, device=device)

# Your sentences
loesungbibliothek_prozessname_text = loesungbibliothek['Prozessname']
# Create embeddings
loesungbibliothek_prozessname_embeddings = model.encode(
    loesungbibliothek_prozessname_text,
    batch_size=32,
    convert_to_numpy=True,
    normalize_embeddings=True
)

print(loesungbibliothek_prozessname_embeddings.shape)  # (3, 384)
```
